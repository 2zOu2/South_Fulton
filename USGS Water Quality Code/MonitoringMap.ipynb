```python
import geopandas as gpd
import matplotlib.pyplot as plt
import json
import plotly.express as px
from shapely import wkt
from shapely.geometry import Point
import pandas as pd
```

```python
gdf = gpd.GeoDataFrame.from_file('City_Limits_2021_20240307.csv')
gdf['geometry'] = gdf['the_geom'].apply(wkt.loads)
gdf.set_geometry('geometry')
```

```python
data = pd.read_csv('pretty_good.csv')
data['ActivityStartDate'] = pd.to_datetime(data['ActivityStartDate'])
data = data[data['ActivityStartDate'] >= '2019-01-01']
```

```python
data
```

```python
geometry = [Point(lon, lat) for lon, lat in zip(data['LongitudeMeasure'], data['LatitudeMeasure'])]
site = gpd.GeoDataFrame(data, geometry=geometry)
```

```python
site['Size'] = 15
```

```python
grouped = site.groupby('OrganizationIdentifier')

for org_id, group_df in grouped:
    choropleth_trace = px.choropleth_mapbox(
        gdf,
        geojson=gdf.geometry,
        locations=gdf.index,
        color_discrete_sequence=["rgba(0,0,0,0)"],
        hover_name='Name',
        title= 'Water Quality Monitering Location Map (Fulton County Only)',
        mapbox_style="carto-positron",
        center={"lat": gdf.geometry.centroid.y.mean(), "lon": gdf.geometry.centroid.x.mean()},
        zoom=8
    )

    choropleth_trace.update_layout(coloraxis_showscale=False)

    scatter_trace = px.scatter_mapbox(group_df, 
                                      lat="LatitudeMeasure", 
                                      lon="LongitudeMeasure", 
                                      hover_name="MonitoringLocationIdentifier",
                                      zoom=10,
                                      size='Size',
                                      color='OrganizationFormalName'
                                     )

    scatter_trace.update_layout(mapbox_style="open-street-map")

    # combined_fig = choropleth_trace.add_trace(scatter_trace.data[0])
    # combined_fig = choropleth_trace.add_trace(scatter_trace.data[1])

    for trace in scatter_trace.data:
        choropleth_trace.add_trace(trace)

    # combined_fig.show()
    
    choropleth_trace.write_html(str(org_id) + "_monitoring_map_after_2019.html")
```
